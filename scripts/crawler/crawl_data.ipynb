```python
person = pd.read_csv("person_2020_update.csv")
person
```

```python
selected_columns = ["name", "gender", "birthyear", "deathyear", "bplace_country", "dplace_country", "age"]
df_selected = person[selected_columns]

df_selected
```

```python
unique_countries = df_selected["bplace_country"].unique()

# Ghi danh sách quốc gia vào file TXT
with open("country.txt", "w", encoding="utf-8") as f:
    for country in unique_countries:
        f.write(str(country) + "\n")
```

```python
with open("countries_asia.txt", "r", encoding="utf-8") as f:
    asian_countries = [line.strip() for line in f]

df_asia = df_selected[df_selected["bplace_country"].isin(asian_countries)]
df_asia = df_asia[(pd.isna(df_asia["deathyear"])) & (df_asia['birthyear'] > 1950)]
df_asia
```

```python
countries = df_asia["bplace_country"].unique()
print(len(countries))
df_asia['bplace_country'].value_counts()
```

```python
df_sampled = df_asia.groupby("bplace_country", group_keys=False).apply(lambda x: x.sample(n=min(40, len(x)), random_state=42))
df_sampled
```

```python
df_sampled.to_csv("people.csv")
```
